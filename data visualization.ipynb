```python
yield_oranges = [0.962, 0.964, 0.567, 0.234, 0.991]
```

```python
plt.xlabel("Years")
plt.ylabel("Yield in tonnes per hectre")
plt.plot(years, yield_apples)
plt.plot(years, yield_oranges)
```

```python
plt.xlabel("Years")
plt.ylabel("Yield in tonnes per hectre")
plt.title("Crop yields")
plt.legend(["apples", "oranges"])
plt.plot(years, yield_apples)
plt.plot(years, yield_oranges)
```

```python
plt.plot(years, yield_apples, marker = 'o', ls = '-', c = 'red', alpha = 0.5, ms = 10)
plt.plot(years, yield_oranges, marker = 'x', ls = '--', c = 'orange', ms = 10)
plt.xlabel("Years")
plt.ylabel("Yield in tonnes per hectre")
plt.title("Crop yields")
plt.legend(["Apples", "Oranges"])
```

```python
plt.plot(years, yield_apples, 'x-b')
plt.plot(years, yield_oranges, 'o--r')
plt.xlabel("Years")
plt.ylabel("Yield in tonnes per hectre")
plt.title("Crop yields")
plt.legend(["Apples", "Oranges"])
```

```python
plt.plot(years, yield_apples, 'x-.b')
plt.plot(years, yield_oranges, 'o--r')
plt.xlabel("Years")
plt.ylabel("Yield in tonnes per hectre")
plt.title("Crop yields")
plt.legend(["Apples", "Oranges"])
```

```python
plt.figure(figsize = (12,6))
plt.plot(years, yield_apples, 'x-.b')
plt.plot(years, yield_oranges, 'o--r')
plt.xlabel("Years")
plt.ylabel("Yield in tonnes per hectre")
plt.title("Crop yields")
plt.legend(["Apples", "Oranges"])
```

```python
setosa_df = df[df.species == 'setosa']
```

```python
virginica_df = df[df.species == 'virginica']
```

```python
plt.title('Distribution Of Sepal Widths')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins = np.arange(2,5,0.25), stacked = True)
plt.legend(['setosa', 'versicolor', 'virginica'])
```

## Plotting multiple charts in a grid

```python
fig,axes = plt.subplots(2,3, figsize = (20,8))
axes[0,0].plot(years, yield_apples, "s-b")
axes[0,0].plot(years, yield_oranges, "o--r")
axes[0,0].set_xlabel("Years")
axes[0,0].set_ylabel("Yield in tonnes per hectre")
axes[0,0].legend(['apples','oranges'])
axes[0,0].set_title("crop Yields")

axes[0,1].set_title("Sepal Length vs sepal Width")
sns.scatterplot(x = df.sepal_length, y = df.sepal_width, hue = df.species, s = 100,ax = axes[0,1])

axes[0,2].set_title("Distribution of sepal width")
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins = np.arange(2,5,0.25), stacked = True)
axes[0,2].legend(['setosa', 'versicolor', 'virginica'])

axes[1,0].set_title("restaurant bills")
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, ax = axes[1,0])

axes[1,1].set_title("Flight Traffic")
sns.heatmap(flights_df, cmap = 'Blues', ax = axes[1,1])

axes[1,2].set_title("Data Science Meme")
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])
plt.tight_layout(pad = 4)
```
